### 5. Comando ".where" y variables del comando ".value_counts"

```python
df_nations['co2_rec'] = np.where(df_nations['co2']> df_nations['co2'].mean(),1,0)
```

```python
df_nations.head()
```

```python
# En este caso, creamos otra columna donde mostrara aquellos paises donde el valor de "Co2" es mayor al promedio, marcando con un "1" a los que si, y un "0" a los que no

# La fórmula quedando de la siguiente manera:

# df_nations['co2_rec'] - Siendo el nombre de la nueva "variable" y "encabezado"

# = np.where() - siendo la operación obtenida mediante la "librería" numpy el cual funciona como la fórmula "if" / "si" de Excel

# (df_nations['co2'] > df_nations['co2'].mean,_,_) - Siendo esta la comparación que debe realizar para aplicar el criterio de "1" o "0"
# en este caso siendo la comparativa de aquellos valores que estén por encima de la media, esto mediante el comando ejecutado por ".mean"

#(_, 1, 0) - Siendo los valores que colocará en la tabla, dependiendo del resultado de la "comparativa" o "rango", anteriormente establecido
```

```python
df_nations['co2_rec'].value_counts()

# Con este comando , podremos ver el total de los diferentes valores que se crearon en la nueva columna "co2_rec", agrupados por los criterios que establecimos
```

```python
df_nations['co2_rec'].value_counts()[1]

# Con este comando podremos ver directamente el valor que cumplan con un criterio; en caso de conocer el valor que estemos buscando entre toda la tabla y simplemente
# queramos conocer el conteo total de las mismas, el valor buscado deberá colocarse OBLIGATORIAMENTE entre corchetes "[]"
```

```python
print('Existen', df_nations['co2_rec'].value_counts()[1], "paises con indices de emisión de co2 mayores al promedio")

# Con este comando podemos darle una presentación más formal al resultado que obtuvimos, en dado caso de necesitar responder a una pregunta

# Igual, como se puede ver, se puede colocar sin ningún tipo de "bug" el comando completo dentro del mismo comando "print"
```
