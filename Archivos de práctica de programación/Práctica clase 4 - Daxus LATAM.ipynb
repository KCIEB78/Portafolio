## **2. Función para crear ciclos / loops mediante "for":**

### *2.1. Ejecución correcta del código:*

```python
numeros = [1, 2, 3, 4, 5]

for i in numeros:
    print(i)
```

En estas líneas de código tenemos lo siguiente:

- **numeros = [1, 2, 3, 4, 5]** - En esta línea se establece una variable de tipo list, que se usará como "rango" y como "los elementos a mostrar"

- **for i in numeros:** - En esta línea se establece la cantidad de ciclos que se llevarán a acabo (un rango).

    - En este caso, se está utilizando la variable previamente definida como la cantidad de ciclos que se llevarán a cabo; siendo que, hay 5 elementos en la lista, por lo que el ciclo se repetirá esas 5 veces
    - Por otro lado, cualquier función escrita después de los dos puntos (:), será aquella que se ejecute por la misma cantidad de ciclos una y otra vez hasta terminar

- **print(i)** - 

Todo esto significa que: la función "print()" se ejecutará **la misma cantidad de veces que la función "for" haya establecido**; de igual manera, se mostrará cada uno de los "resultados" que arroje el ciclo.

En este caso se muestra cada uno de los elementos de la lista, debido a que la función print usó cada uno de los elementos del "rango" establecido por "for" como los elementos que mostraría

### *2.2. Ejecución errónea del código:*

```python
numeros = [1, 2, 3, 4, 5]

for i in numeros:
    print(numeros)
```

En este caso, al momento de ejecutar la función print() **en base a la misma variable**, lo que hizo fue **mostrar la misma lista** definida en la variable, por **la cantidad de veces que se ejecutó el ciclo**, teniendo como resultado la ejecución errónea de la función for

## **3. Función para crear ciclos / loops mediante "while"**

```python
tiempo = 5

while tiempo > 0:
    print(tiempo)
    tiempo = tiempo - 1
```

En estas líneas de código tenemos lo siguiente:

- **tiempo = 5** - En esta línea se establece una **variable de tipo int**, que se utilizará para **limitar los ciclos que se ejecutarán**

- **while tiempo > 0:** - En este línea **establecemos el límite de ciclos que se llevarán a cabo**; en este caso, tenemos que, mientras la variable que establecimos previamente **sea mayor a 0**, entonces el ciclo seguirá ejecutándose. Por otro lado; *así como en la función for*, cualquier otra función escrita **después de los dos puntos (:)** será la que se ejecute *por la misma cantidad de ciclos establecidos*

- **print(tiempo)** - En esta línea establecemos la ejecución de la función print; en este caso, **mostrará el valor de la variable previamente definida** mientras se siga ejecutando el ciclo

- **tiempo = tiempo - 1** - En esta línea estamos reduciendo, **por cada ciclo**, el valor de la variable establecida; de esta forma **limitamos correctamente el ciclo** y evitamos que se **ejecute infinitamente**

## **4. Pruebas lógicas mediante "if":**

```python
lenguaje = "Python"

if lenguaje == "Python":
    print("El mejor lenguaje de programación")
```

```python
lenguaje = "C++"

if lenguaje == "Python":
    print("El mejor lenguaje de programación"),
```

```python
lenguaje = "C++"

if lenguaje != "Python":
    print("No es el mejor lenguaje de programación"),
```

```python
lenguaje = "Python"

if lenguaje != "Python":
    print("No es el mejor lenguaje de programación"),
```

En todos estos ejemplos, tenemos las siguiente líneas:

- **lenguaje = "Python"** - En esta línea se establece una variable de tipo string, el cual contendrá el dato con el que se hará la prueba lógica

- **if lenguaje == / !=** - En esta línea **se está ejecutando la prueba lógica** en donde, si **el resultado de la "comparación" entre ambos valores corresponde con el operador** utilizado para hacer la misma prueba lógica; entonces, el código **procederá con la ejecución de la función escrita después de los dos puntos (:)**
    - En este caso, los operadores utilizados fueron: **"== igual a"** y **"!= diferente a"**

## **5. Ciclos / Loops infinitos:**

```python
while True:
    mensaje = input("Escriba un nombre: ")
    print(mensaje)
```

En estas líneas de código tenemos lo siguiente:

- **while True:** - Esta línea establece un ciclo infinito en donde el limitante es un dato de tipo bool, el cual al no estar alineado con alguna prueba lógica o comparación de dos valores, nunca cambiará o acabará de procesarse

- **mensaje = input("Escriba un nombre: ")** - Esta línea esta definiendo una variable **mediante un dato que ingrese el usuario mediante la función input()**, la cual se seguirá ejecutando mientras el ciclo se repita, permitiendo al usuario ingresar diferentes nombres

- **print(mensaje)** - Esta línea ejecutará la función print(), la cual mostrará el dato ingresado por el usuario mediante la misma variable, siendo diferente cada vez que el usuario ingrese otro dato

Como este bloque de código no tiene algún "punto de ruptura" o algo que "corte el ciclo", este se ejecutará infinitamente, teniendo que parar el programa manualmente y que, en caso de no tener bien configurado el programa Visual Studio Code, puede que te obligue a reiniciar el programa

```python
while True:
    mensaje = input("Escriba un nombre: ")
    print(mensaje)

    if mensaje == "parar":
        break
```

En este caso agregamos las siguientes líneas de código:

- **if mensaje == "parar":** - En esta línea de código establecemos **que si el código que se está ejecutando detecta que el usuario ingreso el dato "parar"**, entonces ejecutará una función escrita **después de los dos puntos (:)**

- **break** - En esta línea de código, simplemente le estamos pidiendo al mismo que: **si la prueba lógica anteriormente establecida se cumple**; **inmediatamente el código se detenga y**, por lo tanto, el ciclo dejará de ejecutarse **a pesar de ser infinito** 

Como podemos ver, en este bloque de código **se agregó una manera manual de detener el ciclo infinito**, mediante **el ingreso de una "palabra clave"**, la cual **inmediatamente terminará cualquier función que se esté ejecutando** en ese momento, **independientemente del tipo de función**

### *6.3. Tercer bloque de código de la práctica:*

```python
import cv2
from cvzone.HandTrackingModule import HandDetector

webcam = cv2.VideoCapture(0)
rastreador = HandDetector(detectionCon = 0.8, maxHands= 2.0)

while True:
    exito, imagen = webcam.read()

    imagen = cv2.resize(imagen, (853, 480))

    coordenadas, imagen_manos = rastreador.findHands(imagen)

    cv2.imshow("Proyecto 4 - IA", imagen)

    if cv2.waitKey(1) != -1:
        break

webcam.release()
cv2.destroyAllWindows()
```

En este tercer bloque de código agregamos la siguiente línea de código:

- **imagen = cv2.resize(imagen, (853, 480))** - Esta línea de código siendo la que nos permite establecer el tamaño de la ventana emergente

En este tercer bloque de código simplemente agregamos la función **para definir el tamaño de la ventana emergente** que se abrirá

### *6.4. Cuarto bloque de código de la práctica:*
