## **3. Limpieza de la base de datos**

## **4. Transformación de variables categóricas**

### *4.1. Creación de la lista de variables categóricas mediante los encabezados de la base de datos*

```python
categoricos = data_clean.select_dtypes(include = ["object"]).columns
```

```python
categoricos
```

```python
lista_categoricos = list(categoricos)
```

```python
lista_categoricos
```

>> Las variables categóricas se refieren a aquellos datos que pueden ser convertidos y reinterpretados mediante el uso de números

>> Por ejemplo: Convertir datos "Si" y "No", a datos "0" y "1"

---

### *4.2. Conversión de los datos de tipo string a variables tipo "dummy"*

```python
data_trans = pd.get_dummies(data_clean, columns = lista_categoricos, drop_first = True)
```

```python
data_trans['Gimnasio_Si'] = data_trans['Gimnasio_Si'].astype(int)
data_trans['Fuma_Si'] = data_trans['Fuma_Si'].astype(int)
data_trans['Nutriólogo_Si'] = data_trans['Nutriólogo_Si'].astype(int)
```

## **7. Creación y evaluación del modelo de datos para la predicción de valores**

### *7.1 Importar las herramientas para la creación del modelo*

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

---

### *7.2. Creación del modelo de datos*

```python
rfr_model = RandomForestRegressor()
```

```python
rfr_model
```

```python
rfr_model.fit(x_train, y_train)
```

---

## **8. Uso y evaluación del modelo de datos para generar predicciones**

### *8.1. Generar las predicciones en base al modelo de datos*

```python
y_train_pred = rfr_model.predict(x_train)
```

```python
y_train_pred
```

```python
y_test_pred = rfr_model.predict(x_test)
```

```python
y_test_pred
```

---

### 8.2. Evaluación de los resultados de la predicción obtenida

#### *8.2.1. Evaluación de los resultados en base a los "datos de entrenamiento"*

```python
train_mse = mean_squared_error(y_train, y_train_pred)
```

```python
train_mse
```

#### *8.2.2. Evaluación de los resultados en base a los "datos de prueba"*

```python
test_mse = mean_squared_error(y_test, y_test_pred)
```

```python
test_mse
```

---

### *8.3. Evaluación del modelo en base a la "regresión lineal"*

```python
train_r2 = r2_score(y_train, y_train_pred)

test_r2 = r2_score(y_test, y_test_pred)
```

```python
train_r2, test_r2
```

## **9. Graficación del modelo de datos**

### *9.1. Grafico del modelo de datos*
