En esta clase se creó un archivo ejecutable de Python que genera un reporte de ciertos datos que introduce el usuario y, a partir de una plantilla, se crea un PDF

## 1. Definir las variables de los datos que nos pide la plantilla

```python
proyecto = input("Escriba la descripción del proyecto: ")

horas_estimadas = input("Escriba las horas estimadas para terminar el proyecto: ")

valor_hora = input("¿Cuál es el valor por hora trabajada del proyecto? ")

plazo_estimado = input("¿Cuánto tiempo cree que tardará el proyecto en desarrollarse? ")

valor_total = int(horas_estimadas) * int(valor_hora)
```

En el caso de la variable "valor_total", se le agrega la función "int" a ambas variables para que solamente contemple los datos de tipo "int" que se introduzcan (en caso de escribir datos como: 28 horas, etc.), y Python pueda realizar correctamente la operación entre ambas variables sin arrojar errores

## 2. Visualizar los datos introducidos por el usuario

```python
print(proyecto)

print(horas_estimadas)

print(valor_hora)

print(plazo_estimado)

print(valor_total)
```

## 4. Ejecutar la función que genera el archivo PDF "base"

```python
pdf = FPDF()
```

## 5. Ejecutar la función que agrega una página en blanco en el PDF y establece la fuente del texto

```python
pdf.add_page()
pdf.set_font("Arial")
```

En este apartado faltaría la función que inserta la "plantilla" del reporte desde la esquina del archivo PDF, siendo esta función:

## 6. Ejecutar las funciones que agrega los datos previamente introducidos por el usuario en la parte de la plantilla donde van los datos

```python
pdf.text(115, 145, proyecto)
pdf.text(115, 160, horas_estimadas)
pdf.text(115, 175, valor_hora)
pdf.text(115, 190, plazo_estimado)
pdf.text(115, 205, str(valor_total))
```

En el caso de la última línea de código, se le agrega la función "str" para convertir el resultado de la operación de las variables previamente realizada, en un dato de tipo "string"; ya que, la función de la librería ".text" solamente permite introducir texto (datos de tipo "string")

Las "coordenadas" deben ser otorgadas por el departamente de diseño o, en dado caso de no tenerlas, utilizar diferentes herramientas para calcular la posición de las mismas

## 7. Ejecutar la función que guarda el archivo PDF en la misma carpeta donde se encuentra el ejecutable una vez haya colocado todos los elementos en su lugar

```python
pdf.output("Presupuesto.pdf")
```

Con esta última línea de código estamos guardando el archivo PDF con la función ".output()", ya con toda la información agregada al mismo y con el nombre establecido entre los paréntesis de la función; en este caso, llamamos al archivo: "Presupuesto"

Una vez ejecutemos el código, el archivo generado se guardará en la misma carpeta donde se encuentre este archivo y con eso termina el código
