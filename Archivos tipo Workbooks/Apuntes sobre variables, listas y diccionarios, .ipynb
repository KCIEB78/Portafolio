Ejemplo: 0.0, 0.10, 124.35, 010, 902.34, etc...

# **2. Tipos de variables en Python para el análisis básico de datos**

## *1.2. Variables en Python:*

### *1.2.1. ¿Cómo declarar una variable en Python?*

#### *1.2.1.1. Declaración de variables:*

Para declarar una variable en Python, *basta con asignarle un **" nombre "** a la variable, seguido del signo **" = "** y finalmente el **valor, texto, oración, etc...***; que uno desea *" almacenar "* en dicha variable.

Teniendo un resultado como: **" variable_n = 1234 ", " variable_t = "número" ", "variable_o = "palabra otra palabra más palabras etc..."**, etc... 

> En el siguiente apartado explico más a detalle como declarar cada tipo de variables

---

> IMPORTANTE: Una vez " declarada una o más variables " en Python, éstas serán guardadas en el mismo archivo HASTA QUE SE CIERRE.

> Por lo que, si uno necesita trabajar otra vez con estas variables al abrir el archivo, se necesita ejecutar otra vez las " líneas de código donde se declararon dichas variables ", lo que permitirá trabajar nuevamente con ellas.

> Pero, en dado caso de que éstas no se ejecuten, Python arrojará un error donde mencionará que dicha variable " no ha sido declarada "; esto, a pesar de que exista la " línea de código " con dicha variable

> Cabe aclarar que en este cuadernillo todavía no entraré en el aspecto de las funciones para visualizar o realizar operaciones entre las variables; por lo que, me enfocaré solamente en como declararlas correctamente y evitar errores en Python

---

#### *1.2.1.2. Límite para el nombre de una variable:*

El " límite de un nombre o dato " para una variable realmente depende de que tipo de dato o información necesites almacenar en la misma.

En el caso del **" límite para el nombre de una variable "**, lo único que hay que tener a consideración, es que **NO SE PUEDEN UTILIZAR ESPACIOS**; por lo que, si necesitas colocar *una o más palabras a dicha variable*, éstas **OBLIGATORIAMENTE** necesitan estar separadas por un " guión bajo _ ". Esto, para evitar que Python arrojé un error y nos permita almacenar dicha variable.

Ejemplo : " variable_de_ejemplo_para_el_límite = ejemplo "

```python
variable_de_ejemplo_para_el_límite = "ejemplo"
```

---

> IMPORTANTE: En el caso del límite para las variables, depende del " tipo de dato " que se quiera almacenar en las mismas; por lo que, esta parte la abarcaré en los apartados de cada tipo de dato

#### *1.2.1.3. ¿Cómo saber si mi variable se guardó correctamente?*

Para saber *si la variable declarada se guardó correctamente*, basta con ejecutar la **" línea de código "** y asegurarnos de que debajo de la misma variable aparezca un **" checkmark "** junto al número *que indica el tiempo que tardó la misma línea para ejecutarse*.

Si al ejecutarse dicha " línea de código " Python no muestra ningún error, significa que la variable a sido declarada correctamente y ya puede utilizarse en el archivo.

#### *1.2.1.4. ¿Se pueden declarar varias variables con el mismo nombre?*

En el " ambiente de programación de Python " refiriéndonos al análisis de datos, éste si permite crear muchas variables asignándoles el mismo nombre; pero, en la práctica esto resulta contraproducente en dado caso de ejecutarlas o necesitarlas para realizar operaciones o ejecutar diferentes funciones

## *1.2. Operadores para comparar valores o variables*

### *1.2.1. Descripción de los operadores:*

Los diferentes signos que Python acepta para poder generar comparación entre 2 o más valores o variables de CUALQUIER TIPO DE DATO, son:

- " > " Operador que hace referencia a la comparación " Mayor que ", usado como: " a es mayor que b "

- " < " Operador que hace referencia a la comparación " Menor que ", usado como: " a es menor que b "

- " >= " Operadores que hacen referencia a la comparación " Mayor o igual que ", usados como: " a es mayor o igual que b "

- " <= " Operadores que hacen referencia a la comparación " Menor o igual que ", usados como: " a es menor o igual que b "

- " != " Operadores que hacen referencia a la comparación " Diferente a / que ", usados como: " x es diferente a / que y "

- " == " Operadores que hacen referencia a la comparación " Igual a / que ", usados como: " x es igual a / que y "

---

> IMPORTANTE: Al momento de usar estos operadores para hacer comparaciones entre valores o variables de cualquier tipo de dato, recomiendo encarecidamente decidir desde un inicio si es que se van a ocupar tíldes, simbolos especiales como: " # , $ , & , etc... ", para la declaración de variables o en el texto que se ocupará en la documentación, base de datos, etc...

> Esto debido a la existencia de los términos " case sensitivity " y " coincidencia exacta del identificador "; los cuales son los que permiten que dichas variables sean " unicas ". Por lo que, si uno quiere " llamarlas " en otra parte del documento, SE DEBERÁN ESCRIBIR EXACTAMENTE IGUAL a como fueron declaradas dichas variables.
