## *1.1. Funciones ejecutables - Explicación*

### *1.1.1. ¿Qué son las funciones ejecutables en Python?*

Las funciones ejecutables en Python son palabras que forman parte de una lista extensa que, al momento de escribirlas, llaman al entorno de programación en Python para realizar una acción en específico

Algunas funciones pueden ser llamadas mediante el uso de un **punto "."** o directamente escribiéndola; al igual que, algunas funciones requieren del uso de **paréntesis ()** para que Python pueda ejecutarlas correctamente

### *1.1.2. ¿A qué hace referencia la palabra "sintaxis" cuando se habla de funciones?*

Así como en programas como *Microsoft Excel*, las funciones que se utilizan se basan en una **"sintaxis"** que define el orden en que **OBLIGATORIAMENTE se tienen que introducir / escribir los datos**, para que dichas funciones, al momento de ser ejecutadas en una línea de código, den el resultado deseado y no arrojen errores de ningún tipo.

Por lo que, **conocer las sintaxis de cada una de las funciones permite entender su comportamiento** y, *en caso de querer usarlas en otros contextos o empezar a hacer combinaciones extensas de funciones*, uno pueda **encontrar y solucionar cualquier error** que arroje la función **con tan solo volver a leer la escritura del código** sin depender de la descripción del error que muestra Python en dado caso de haberlo

### *1.1.3. Al momento de documentar una línea de código, ¿hay alguna manera de escribir un texto sin la necesidad de hacer uso de citas de código o celdas tipo Markdown?*

#### *1.1.3.1. Comentarios en celdas de tipo Code:*

En caso de querer documentar y explicar una línea de código sin la necesidad de usar varias celdas de tipo Markdown o realizar citas de código, basta con que, en la misma celda de tipo Code, se agregue un hash (#) antes del texto que se quiera escribir dentro de la misma.

Esto creará un "comentario" que permite escribir líneas del código que no influirán en ningún aspecto del mismo; pero que, OBLIGATORIAMENTE, debe agregarse un hash (#) por cada línea de código que se desee escribir como comentario, en caso de agregar uno en muchas líneas de código (lo cual no es una buena práctica)

De igual manera, se puede hacer uso de las triples comillas o tildes (que menciono en el cuadernillo: "Apuntes sobre variables, listas y diccionarios"); pero, cabe recordar que, de todas formas, Python interpretará estas líneas como un dato de tipo string

Por lo tanto, hay que tener cuidado si uno desea agregar comentarios utilizando este último método

## *1.2. Funciones ejecutables - Explicación y ejemplos:*

### *1.2.1. Función "print"*

#### *1.2.1.1. Descripción de la función:*

La función print es aquella que:

- No requiere de un punto "." para ser llamada

- Requiere del uso de paréntesis para poder ejecutarse correctamente

- Nos permite visualizar en la consola o debajo de la celda de tipo código cualquier tipo de dato o variable previamente definida, que se escriba dentro de los paréntesis

#### *1.2.1.2. Ejemplos en celdas de tipo Code*

##### *1.2.1.2.1. Ejemplos con diferentes tipos de datos:¨*

```python
print(5)

# Este ejemplo siendo un dato de tipo int
```

```python
print(350.23)

# Este ejemplo siendo un dato de tipo float
```

```python
# Para este primer ejemplo ejecutaremos una de las primers funciones que uno aprende al momento de adentrarse en el mundo de la programación

print("Hola Mundo")

# Este primer ejemplo siendo un dato de tipo string

# Como se podrá ver en el resultado, Python nos mostró lo que escribimos dentro del paréntesis; en caso de escribir la línea de código en un archivo ".py"
# este abrirá una terminal de CMD donde se mostrará el mismo mensaje 
```

```python
print(125 < 800)

# Este ejemplo siendo un dato de tipo bool

# En este caso, la función print nos mostrará el resultado de la comparación de ambos valores en vez de los datos que se hayan escrito
```

```python
print(["Lista",
       "creada",
       "como",
       "ejemplo"])

# Este ejemplo siendo un dato de tipo list
```

```python
print({
    "Entrada 1" : "Información gramatical",
    "Entrada 2" : "Información gramatical",
    "Entrada 3" : "Información gramatical"
})

# Este ejemplo siendo un dato de tipo dict
```

#### *1.2.1.3. Sintaxis de la función

La sintaxis de la función print() es la siguiente

- La palabra "print" llama a la función que nos permite mostrar el dato deseado

- Dentro del paréntesis se escribe el dato (siguiendo las reglas de cada tipo) o variable que se desee mostrar
