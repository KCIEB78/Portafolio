# **1. Funciones ejecutables - Explicación**

## *1.1. ¿Qué son las funciones ejecutables en Python?*

Las funciones ejecutables en Python son palabras que forman parte de una lista extensa que, al momento de escribirlas, llaman al entorno de programación en Python para realizar una acción en específico; teniendo en cuenta que:

- Algunas funciones pueden ser llamadas mediante el uso de un **punto "."** o directamente escribiéndolas al inicio de la línea de código

- La mayoría de funciones requieren del uso de **paréntesis ()** para que Python pueda ejecutarlas correctamente o permita ver el resultado de ejecutar las mismas

>> Mediante la explicación de las sintaxis de cada una de las funciones, desarrollaré esta parte de los elementos que las compondren

## *1.2. ¿A qué hace referencia la palabra "sintaxis" cuando se habla de funciones?*

Así como en programas como *Microsoft Excel*, las funciones se basan en una **"sintaxis"** que define el orden en que **OBLIGATORIAMENTE se tienen que escribir los elementos que requieren las mismas para ser ejecutadas correctamente (incluyendo el mismo nombre de la función)**; de esta manera, al momento de ser ejecutadas en un bloque o línea de código, den el resultado deseado y no arrojen errores.

De igual manera, quiero **resaltar la importancia de conocer las sintaxis de cada una de las funciones**; ya que, esto nos permite entender su comportamiento **en caso de querer usarlas en otros contextos o empezar a hacer combinaciones extensas de funciones**. De igual manera, esto nos permite  **encontrar y solucionar cualquier error** que arroje la función **con tan solo volver a leer la escritura del código** *sin depender de la descripción del error que muestra Python*

## *2.2. Función "f-string (f'' / f"" / f''' _ ''' / f""" _ """)"*

### *2.2.1. Descripción de la función:*

La función f' string es aquella que nos permite crear un dato de tipo string; pero que, **se le pueden agregar variables en medio del mismo, de cualquier tipo**; además, **permite el uso de las triples comillas o tildes** para separar en diferentes líneas el texto escrito

### *2.2.2. Ejemplos en celdas de tipo Code:*

```python
j = "27 palabras"
k = 9780
l = 9779.91
m = ["Redondeo", "siguiendo sus reglas"]
n = {"Redondeo" : "Reducir", "Reglas" : "Normas"}

f'En cuestión de {j} explico como {l} se convierte en {k} mediante el {m}, teniendo en cuenta los dos principios: {n}'

# Ejemplo de la aplicación de la función f-string, insertando diferentes variables de distintos tipos entre un texto de una sola línea de código
```

```python
o = 10000 - 6743.12
p = 5476.24
q = o + p
r = "operaciones básicas entre variables"
s = ["suma", "resta", "multiplicación", "división"]
t = {"Norma 1" : "Jerarquía de operaciones", "Norma 2" : "Redondeo de números decimales"}
u = f'la suma de {o} y {p} da el resultado de {round(q,2)}'
v = f'''{u}, siendo ésta un ejemplo sobre las {r},
teniendo en cuenta que estas son: {s} 
y las normas de {t}
'''

print(v)

# Ejemplo de la aplicación de la función f-string, insertando diferentes variables de distintos tipos entre un texto en diferentes líneas de código

# De igual manera, se puede observar que la función también puede ser ejecutada para definir una variable
```

### *2.2.3. Sintaxis de la función*

#### *2.2.3.1. Sintaxis de la función en la escritura de un dato de tipo string:*

La sintaxis de la función f-string para la escritura de un dato de tipo string, es la siguiente:

1. **f** - La "letra f" es la que nos permite **llamar a la función**

2. **' ' / " " / ''' ''' / """ """** - Las tildes o comillas **funcionan como las de un dato tipo string**, teniendo que **escribir el texto en medio de éstas**

3. **{Nombre de la variable definida}** - Entre el **par de llaves ({})**, se deberá escribir **el nombre de la variable definida de cualquier tipo**, *estas pueden ser escritas en cualquier parte del texto*, **siempre y cuando estén dentro de las comillas o tildes**

#### *2.2.3.2. Sintaxis de la función en la escritura de una variable de tipo string:*

La sintaxis de la función f-string es la siguiente:

1. **(Nombre de la variable definida) =** - La escritura del **nombre de una variable y el signo de igual (=)** nos permite **guardar un texto generado mediante la función f-string**

2. **f** - La "letra f" es la que nos permite **llamar a la función**

3. **' ' / " " / ''' ''' / """ """** - Las tildes o comillas **funcionan como las de un dato tipo string**, teniendo que **escribir el texto en medio de éstas**

4. **{Nombre de la variable definida}** - Entre el **par de llaves ({})**, se deberá escribir **el nombre de la variable definida de cualquier tipo**

## *5.3. Función para ejecutar ciclos mediante varios métodos (while "")*

### *5.3.1. Descripción de la función (while ""):*

La **función while** es la que nos permitirá ejecutar un ciclo mediante varios métodos, estos pueden ser:

- La ejecución de un ciclo mediante un dato de tipo bool; mientras el dato se identifique como "True", hasta que se identifique como "False" y viceversa

>> Este método puede crear ciclos infinitos; por lo que, hay que tener cuidado al momento de usar este método al momento de usar la función while

- La ejecución de un ciclo mediante un número; mientras el número mantenga el mismo valor o hasta que el número se encuentre dentro de un rango (esto mediante el uso de los operadores para realizar comparaciones: mayor que >, menor que <, igual a =, etc.)
