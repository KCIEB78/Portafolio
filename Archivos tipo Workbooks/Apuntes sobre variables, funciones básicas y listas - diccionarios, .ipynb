# **1. Tipos de datos  y variables en Python**

## *1.1. Tipos de datos:*

### *1.1.1. Tipo de dato " int ":*

Este tipo de dato hace referencia a todos aquellos que son **" números enteros "** que NO TIENEN DECIMALES, *incluyendo el decimal " .0 "*

Una vez *se guarde una variable*, o que *el elemento de una lista contenga un número entero*; Python automáticamente lo guardará y clasificará como " int "

### *1.1.2. Tipo de dato " float ":*

Este tipo de dato hace referencia a todos aquellos que son **" números con " n " cantidad de decimales "**, *incluyendo el decimal " .0 "*

Una vez *se guarde una variable*, o que *el elemento de una lista contenga un número con cualquier decimal (0.1, 0.02, 0.320, etc...)*, Python automáticamente lo guardará y clasificará como **" float "**

### *1.1.3. Tipo de dato " string ":*

Este tipo de dato hace referencia a todos aquellos que contienen una **" Letra "**, **" Palabra "** o una **" Cadena de Texto "**  , incluyendo **" Texto con números "** o un **" Espacio en blanco "**

>> No se recomienda guardar letras como variables, al menos que necesites guardar un símbolo o viñeta especial, como los que tengo anotados en el archivo: " Apuntes sobre bloques de textos " o uno que no puedas escribir en tu teclado de forma manual (por codigo ASCII o la combinación de " CTRL " o " ALT " y cualquier tecla o tecla del pad numérico)

Una vez *se guarde una variable que contenga cualquiera de los elementos previamente mencionados*, Python automáticamente lo guardará y clasificará como **" string "**

### *1.1.4. Tipo de dato " bool ":*

Este tipo de dato hace referencia a aquellos que contienen datos que se pueden clasificar así mismos como **" verdadero "** o **" falso "**

Este tipo de dato puede *ser guardado mediante la declaración de una variable* donde se ejecuten funciones que generan **" consultas "** o **" comparativas "**; en ese caso, Python guardará dicha variable y clasificará el dato almacenado en ella como " bool "

## *1.2. Operadores para comparar valores o variables*

### *1.2.1. Descripción de los operadores:*

Los diferentes signos que Python acepta para poder generar comparación entre 2 o más valores o variables de CUALQUIER TIPO DE DATO, son:

- " > " Operador que hace referencia a la comparación " Mayor que ", usado como: " a es mayor que b "

- " < " Operador que hace referencia a la comparación " Menor que ", usado como: " a es menor que b "

- " >= " Operadores que hacen referencia a la comparación " Mayor o igual que ", usados como: " a es mayor o igual que b "

- " <= " Operadores que hacen referencia a la comparación " Menor o igual que ", usados como: " a es menor o igual que b "

- " != " Operadores que hacen referencia a la comparación " Diferente a / que ", usados como: " x es diferente a / que y "

- " == " Operadores que hacen referencia a la comparación " Igual a / que ", usados como: " x es igual a / que y "

> IMPORTANTE: Al momento de usar estos operadores para hacer comparaciones entre valores o variables de cualquier tipo de dato, recomiendo encarecidamente decidir desde un inicio si es que se van a ocupar tíldes para la declaración de variables o en el texto en general que se ocupará en la documentación, base de datos, etc...

> Ya que, existen los términos "case sensitivity " y " coincidencia exacta del identificador "; los cuales son los que permiten que dichas variables sean " unicas " y que requieren OBLIGATORIAMENTE que se escriban dichas variables para " llamarlas ", tal cual se declararon desde un inicio

> Inclusive al momento de ya estar trabajando con tablas se puede ver perfectamente como estos dos elementos influyen cuando se quieren generar filtros y otras técnicas que se pueden ejecutar en tablas o bases de datos
