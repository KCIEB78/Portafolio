## *1.6. División de "piso" " / / "*

**Descripción:**

Las divisiones de piso se refieren a divisiones que, aunque se usen números enteros o decimales, siempre dará un resultado como un " número entero " sin excepeción; estas, se realizan mediante la escritura de dos " / " en medio de los valores.

**Comportamiento del redondeo:**

- Si el resultado queda en un número con decimal **" 0.99 "**, el redondeo siempre se realizará hacia la " derecha ", quedando como **" 1.00 "**
- Si el resultado queda en un número con cualquier decimal antes de  **" 0.99 " (0.98, 0.97, 0.96, 0.95, 0.94, 0.93, 0.92, 0.91, 0.90, ...)**, el redondeo **SIEMPRE** se realizará hacia la " izquierda ", quedando como **" 0.00 "**

```python
80 // 5
```

```python
700.14 // 50.12
```

```python
8000 // 24 // 36 // 7
```

```python
100000 // 241.5 // 842.7 // 96.1
```

# *1.7. *División de restante " % " y de piso restante " %% "* 

**Descripción:**

Las divisiones de " restante " se refieren a divisiones que, a pesar de seguir el mismo comportamiento que una división " normal " y una división " de piso "; siempre se mostrará el " residuo " de dichas divisiones

> En todos estos tipos de operaciones, la manera en que Python las " ejecutará " siempre será de izquierda a derecha
